```python
import os, string, glob
from segtok.tokenizer import web_tokenizer, split_contractions
```

```python
def simple_filter(word):
    term = word.lower()
    for p in string.punctuation:
        term = term.replace(p, ' ')
    term = ' '.join([ w for w in split_contractions(web_tokenizer(term)) ])
    return term.strip()
```

```python
def readfile(filepath):
    with open(filepath, encoding='utf8') as infile:
        content = infile.read()
    return content
def get_datasetid(datasetdir):
    return datasetdir.rsplit(os.path.sep,2)[1]
def get_appname(outputdir):
    return '_'.join([ config for config in os.path.dirname(outputdir).split(os.path.sep)[-2:] if config != 'None'])
def get_docid(dockeypath):
    return os.path.basename(dockeypath).replace('.txt','').replace('.key','').replace('.out','').replace('.phrases','')
```

```python
def build_ground_truth(datasetdir, listoffilters=[simple_filter]):
    keysfiles = glob.glob(os.path.join(datasetdir, 'keys','*'))
    qrels = {}
    for i, keyfile in enumerate(keysfiles):
        docid = get_docid(keyfile)
        gt = {}
        keysunfiltered = readfile(keyfile).split('\n')
        for goldkey in keysunfiltered:
            for termfilter in listoffilters:
                goldkey = termfilter(goldkey)
            if goldkey not in gt and len(goldkey) > 0:
                gt[goldkey] = ('k%d' % len(gt), True)
        qrels[docid] = gt
        print('\r%.2f' % (100.*i/len(keysfiles)), end='% ')
    print('\r%.2f' % (100.*(i+1)/len(keysfiles)))
    return qrels
        
```

```python
def nonNumericList(listofkeys):
    return [ (100./(1.+i), kw) for i, kw in enumerate(listofkeys) if len(kw) > 0 ]
def sortedNumericList(listofkeys):
    toreturn = []
    for key in listofkeys:
        if len(key) > 0:
            kw, weight = key.rsplit(' ', 1)
            toreturn.append( (float(weight), kw) )
    return toreturn

def getconversor(datasetid):
    if datasetid.startswith('Rake'):
        return sortedNumericList
    return nonNumericList

def exactmatch(kw, gt):
    # Exact match
    return kw in gt

def ingroundtruth(kw, gt):
    # Considering if the candidate keyphrase is inside some gold keyphrase
    for goldkey in gt:
        if kw in goldkey:
            return True
    return False

def inkeyphrase(kw, gt):
    # Considering if some gold keyphrase is inside the candidate keyphrase  
    for goldkey in gt:
        if goldkey in kw:
            return True
    return False
        
def inoutgroundtruth(kw, gt):
    return inkeyphrase(kw, gt) or ingroundtruth(kw, gt)
```

```python
def build_result(qrels, outputdir, datasetid,
                 listoffilters=[simple_filter],
                 isthekeygold=exactmatch,
                 converter=nonNumericList):
    listofresults = glob.glob(os.path.join(outputdir, datasetid, '*'))
    toreturn = []
    
    for i, resultdoc in enumerate(sorted(listofresults)):
        docid = get_docid(resultdoc)
        gt = qrels[docid]
        seen = set()
        result=[]
        for weight, kw in converter(readfile(resultdoc).split('\n')):
            for termfilter in listoffilters:
                kw = termfilter(kw)
            if len(kw) == 0:
                continue
            if kw not in gt:
                isrel = isthekeygold(kw, gt)
                idkw = ('%sk%d' % ('' if isrel else 'u', len(gt))) #uK%d if unrel key or K%d to rel new keys
                gt[kw] = (idkw, isrel)
            idkw, isrel = gt[kw]
            if idkw not in seen:
                seen.add(idkw)
                result.append( idkw )
        qrels[docid] = gt
        toreturn.append( (docid, result) )
        print('\r%.2f' % (100.*i/len(listofresults)), end='% ')
    print('\r%.2f%%' % (100.*(i+1)/len(listofresults)))
    return ( qrels, toreturn )
        
        
```

```python
outputdirs = ['../../../output/Rake/None/', '../../../output/KEA/NoStemmer/', '../../../output/KEA/none/']
datasetdir = '../../../Datasets/Inspec/'
listoffilters=[simple_filter]
functiontomatch=exactmatch
datasetid = get_datasetid(datasetdir)
print("Generating ground truth")
qrels = build_ground_truth(datasetdir)
for outputdir in outputdirs:
    appname = get_appname(outputdir)
    conv = getconversor(appname)
    print("Generating result ID to %s" % appname)
    qrels, toreturn = build_result(qrels, outputdir, datasetid, listoffilters=listoffilters, isthekeygold=functiontomatch, converter=conv)
    for docid, results in toreturn:
        for i, key in enumerate(results):
            print(docid, 'Q0', key, i, (len(results)-i), appname)
```
